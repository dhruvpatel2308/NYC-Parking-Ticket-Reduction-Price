```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

# Load your dataset
data = pd.read_csv('parking_violation_geocoded.csv')
data = data.drop('unregistered_vehicle',axis=1)
numerical_variables = data.select_dtypes(include=[int,float]).columns
# Select the relevant numerical variables
#numerical_variables = ['fine_amount', 'var1', 'var2', 'var3', ...]  # Replace var1, var2, var3, ... with your variable names
nan_df = data.columns[data.isna().any()].tolist()
print(nan_df)
print("******************************")
data_filled = data.fillna(data.mean())
print("******************************")
# Calculate the correlation matrix
correlation_matrix = data_filled[numerical_variables].corr()
print("******************************")
# Display the correlation matrix
print(correlation_matrix)

# Select the variables with high correlation with fine_amount
selected_variables = ['TotalAmount','law_section','violation_code','lat','lon','vehicle_year','violation_precinct','issuer_precinct']  # Replace with your selected variables
print("******************************")
# Perform multiple linear regression
X = data_filled[selected_variables]
X = sm.add_constant(X)  # Add a constant (intercept) to the model
y = data_filled['fine_amount']

print("******************************")
model = sm.OLS(y, X).fit()
print("******************************")
# Print the regression summary
print(model.summary())
print("******************************")

# Calculate VIF to check for multicollinearity
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif["Variable"] = X.columns
vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

# Display the VIF values
print(vif)
```
